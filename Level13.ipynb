## Bandit Level 13 → Level 14

The password for the next level is stored in /etc/bandit_pass/bandit14 and can only be read by user bandit14. For this level, you don’t get the next password, but you get a private SSH key that can be used to log into the next level. Look at the commands that logged you into previous bandit levels, and find out how to use the key for this level.

- Loggin in, we are provide with the `sshkey.private`

- The instruction said that we need to use this private key in order to log into the next level

```python
ssh -i sshkey.private -p 2220 bandit14@bandit.labs.overthewire.org
```

- However, it won't let us login easily and presented us with the following errors:

```python
Could not create directory '/home/bandit13/.ssh' (Permission denied).
Failed to add the host to the list of known hosts (/home/bandit13/.ssh/known_hosts).

!!! You are trying to log into this SSH server with a password on port 2220 from localhost.
!!! Connecting from localhost is blocked to conserve resources.
!!! Please log out and log in again.
```

- So, we need a workaround through this by copying the private key into our local machine. But first, we need to log out, and execute the `scp` command

```python
scp -P 2220 bandit13@bandit.labs.overthewire.org:~/sshkey.private .
```

- After executing `ls`, we can now see that the private key is now in our local machine. We can try logging in into next level using the private key

```python
ssh -i sshkey.private -p 2220 bandit14@bandit.labs.overthewire.org
```

- However, there is still a problem. 

```python
@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
@         WARNING: UNPROTECTED PRIVATE KEY FILE!          @
@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
Permissions 0640 for 'sshkey.private' are too open.
It is required that your private key files are NOT accessible by others.
This private key will be ignored.
Load key "sshkey.private": bad permissions
```

- A workaround for this would be changing the permission of the private key. First, we need to know the current permissions of the file (by executing `ls -l`) and we can use `chmod` to change this

```python
# Output of ls -l

-rw-r----- 1 five five 1679 Jan 21 10:16 sshkey.private
```

- From the output, we can see that we have read and write permissions but we don't have execute. 

```python
chmod 700 sshkey.private
```

- Running `ls -l` again would reflect that we can execute the file

```python
-rwx------ 1 five five 1679 Jan 21 10:16 sshkey.private
```

- Let's trying logging in again

```python
ssh -i sshkey.private -p 2220 bandit14@bandit.labs.overthewire.org
```

- Now, we can log into `bandit14`

- As said in the instruction, the password for the `bandit14` is located in `/etc/bandit_pass/bandit14`

```python
cat /etc/bandit_pass/bandit14
```
