The password for the next level is stored in the file data.txt, where all lowercase (a-z) and uppercase (A-Z) letters have been rotated by 13 positions

- Reading the content of the `data.txt` gives us a gibberish output. The data is encrypted and we need to use a command that decrypts it (like Ceasar's cipher). We can use the `tr` command for this

```python
cat data.txt | tr 'A-Za-z' 'N-ZA-Mn-za-m'
```

- The first argument of the `tr` commands defines the input range. It tells us that the said command looks for every letter in the alphabet regards of its case. The second argument on the other hand defines the output mapping. Since it has been rotated by 13 positions, we will start with the 14th letter in the alphabet (N) to Z. Then followed by the rest of the alphabet. The second part is for the lower case. Combining `cat` and `tr` through pipe will produce the desired output

```python
The password is 7x16WNeHIi5YkIhWsfFIqoognUTyj9Q4
```
