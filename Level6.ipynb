## Bandit Level 6 → Level 7

The password for the next level is stored somewhere on the server and has all of the following properties:

- owned by user bandit7
- owned by group bandit6
- 33 bytes in size

- As shown in the previous challenge, the `find` command is very useful and powerful espescially when there are many directories and files present. Now, we'll use it again using the parameters above

```python
find / -user bandit7 -group bandit6 -size 33c 2>/dev/null
```

Some takeaways from the command above:

- / → Searches the entire file system starting from the root
- 2>/dev/null → Hides all the "Permission denied" errors

- Now, it presents us with the location of the file that mathches the parameters given

```python
/var/lib/dpkg/info/bandit7.password
```

- All we need to do now is to read its content, and ta-da, we have the password for the next level:

```python
morbNTDkSW6jIlUc0ymOdMaLnOlFVAaj
```
