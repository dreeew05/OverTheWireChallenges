- Now, we need to convert that hexdump into its original format

```python
xxd -r data.txt > file_data
```

- Upon executing the `ls` command, we can see that that there is a file called `file_data` created

```python
# Output of ls -l command

-rw-r----- 1 bandit12 bandit12 2581 Jan 20 06:42 data.txt
-rw-rw-r-- 1 bandit12 bandit12  605 Jan 20 06:46 file_data
```

- Since it was said that the file has been repeated compressed, we need to find a way to uncompress it to its original state. To know which uncompression command to use, we need to determine first what is its type of file

```python
file file_data
```

```python
file_data: gzip compressed data, was "data2.bin", last modified: Tue Oct 14 09:26:00 2025, max compression, from Unix, original size modulo 2^32 572
```

- Apparently, it is a gzip compressed data. Thus, we need to use `gunzip` to uncompress it. However, we need to put a file extension first since executing it without the extension will result to an error

```python
mv file_data file_data.gz
```

```python
# Output after renaming

data.txt  file_data.gz
```

```python
gunzip file_data.gz
```

```python
# Output after extracting

data.txt  file_data
```

- Executing `ls` will still give us the same files in the directory. However, executing the `file` command again to determine the file type of the `file_data` yields a bzip2 file

```python
# Output of file file_data

file_data: bzip2 compressed data, block size = 900k
```

- Now we need to repeat the process again but this time, it would be in `bunzip2` format

```python
mv file_data file_data.bz2
```

```python
# Output after renaming

data.txt  file_data.bz2
```

- After running the `ls`, no additional files were added. However, running the `file` command on `file_data` reveals that it is in the gunzip format. So again, we need to repeat the same process to extract the file

- After doing that shit again, determining the file type reveals that it is now a `POSIX tar archive` (finally!)

```python
# Output of file file_data

file_data: POSIX tar archive (GNU)
```

- Now, let's uncompress it

```python
tar -xf file_data
```

- After executing the `ls` command, we can see that there is a new file created

```python
# Output of ls -l

-rw-r--r-- 1 bandit12 bandit12 10240 Oct 14 09:26 data5.bin
-rw-r----- 1 bandit12 bandit12  2581 Jan 20 06:42 data.txt
-rw-rw-r-- 1 bandit12 bandit12 20480 Jan 20 06:46 file_data
```

- Using the `file` command, the `data5.bin` appears to be a `POSIX tar archive`. Thus, we need to execute the previous decompression command again for this file.

```python
tar -xf data5.bin
```

- It appears that it created another file

```python
# Output of ls -l

-rw-r--r-- 1 bandit12 bandit12 10240 Oct 14 09:26 data5.bin
-rw-r--r-- 1 bandit12 bandit12   219 Oct 14 09:26 data6.bin
-rw-r----- 1 bandit12 bandit12  2581 Jan 20 06:42 data.txt
-rw-rw-r-- 1 bandit12 bandit12 20480 Jan 20 06:46 file_data
```

- The new file `(data6.bin)` appears to be bzip2 file. So, we need to uncompress it again! But first, we need to change the file extension

```python
mv data6.bin data6.bz2
```

```python
bunzip2 data6.bz2
```

- Executing it will not result into a new file. But surprise, surprise, the file type changed again into a `POSIX tar archive`

```python
tar -xvf data6.bin
```
