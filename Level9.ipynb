## Bandit Level 9 → Level 10

The password for the next level is stored in the file data.txt in one of the few human-readable strings, preceded by several ‘=’ characters.

- This is a tricky one NGL since we are dealing with mixed characters that can't be processed normally through grep. Thus, we need to clean it first by filtering only the human-readable strings and piping the `grep` command to find what we are looking for

```python
strings data.txt | grep -E "==+"
```

- The `-E` flag is regex. `==+` means to find the string that has a prefix of multiple `=`. Upon executing the command, we are presented by this result

```python
========== the
========== password
E========== is
5========== FGUW5ilLVJrxX9kMYMmlN4MgbpfMiqey
```

Clearly, the password is

```python
FGUW5ilLVJrxX9kMYMmlN4MgbpfMiqey
```
