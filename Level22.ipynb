- Now, let's see the content of the cronjob for this level

```python
bandit22@bandit:~$ cat /etc/cron.d/cronjob_bandit23
@reboot bandit23 /usr/bin/cronjob_bandit23.sh  &> /dev/null
* * * * * bandit23 /usr/bin/cronjob_bandit23.sh  &> /dev/null
```

- Same with the previous level, we need to read the contents of the bash script

```python
bandit22@bandit:~$ cat /usr/bin/cronjob_bandit23.sh
#!/bin/bash

myname=$(whoami)
mytarget=$(echo I am user $myname | md5sum | cut -d ' ' -f 1)

echo "Copying passwordfile /etc/bandit_pass/$myname to /tmp/$mytarget"

cat /etc/bandit_pass/$myname > /tmp/$mytarget
```

- Now, it is getting tricky here. Analyzing the script would lead us to getting the password for the current user `(bandit22)`. However, we don't want that as we want the password for the next level. Thus, we need to execute the similar command in the terminal (since writing files is not permitted in the current directory)

```python
bandit22@bandit:~$ bash << 'EOF'
> myname="bandit23"
> mytarget=$(echo I am user $myname | md5sum | cut -d ' ' -f 1)
> echo "Copying passwordfile /etc/bandit_pass/$myname to /tmp/$mytarget"
> EOF
```

- With this, we can insert multiple lines until `EOF` is present. We just changed the value of `myname` so that it will get the password for the next level. The output for it is:
